```python
import networkx as nx
import heapq
import time
from scipy.io import mmread
import matplotlib.pyplot as plt


```

==========================
Load Graph
==========================


```python

def load_graph(file_path):

    if file_path.endswith(".mtx"):

        matrix = mmread(file_path)

        G = nx.from_scipy_sparse_array(matrix)

    elif file_path.endswith(".edges"):

        G = nx.Graph()

        with open(file_path, "r") as file:

            for line in file:

                if line.startswith("%"):
                    continue

                parts = line.split()

                if len(parts) >= 2:

                    u = int(parts[0])
                    v = int(parts[1])

                    G.add_edge(u, v)

    else:
        raise ValueError(f"Unsupported file format: {file_path}")

    return G


```

==========================
Hub Labeling (Simplified GPS-style)
==========================


```python

class HubLabeling:

    def __init__(self, graph, max_hubs=10):

        self.graph = graph
        self.max_hubs = max_hubs

        # each node stores: {hub: distance}
        self.labels = {}

        # selected hubs (highest degree nodes = realistic heuristic)
        self.hubs = sorted(
            graph.nodes(),
            key=lambda x: graph.degree(x),
            reverse=True
        )[:max_hubs]

```

==========================
PREPROCESSING
==========================


```python
    def preprocess(self):

        for hub in self.hubs:

            distances = nx.single_source_dijkstra_path_length(
                self.graph,
                hub
            )

            self.labels[hub] = distances


```

==========================
QUERY (GPS FAST MODE)
==========================


```python
    def query(self, source, target):

        best = float("inf")

        for hub in self.hubs:

            if (
                hub in self.labels
                and source in self.labels[hub]
                and target in self.labels[hub]
            ):

                cost = (
                    self.labels[hub][source]
                    + self.labels[hub][target]
                )

                best = min(best, cost)

        return best


```

==========================
DATASETS
==========================


```python

datasets = [
    "inf-USAir97.mtx",
    "inf-power.mtx",
    "road-roadNet-CA.mtx",
    "road-euroroad.edges"
]


```

MAIN

```python

if __name__ == "__main__":

    print("\n===== HUB LABELING RESULTS =====\n")

    dataset_names = []
    preprocessing_times = []
    query_times = []
    node_counts = []
    edge_counts = []

    for dataset in datasets:

        print("=" * 60)
        print("Dataset:", dataset)

        G = load_graph(dataset)

        print("Nodes:", G.number_of_nodes())
        print("Edges:", G.number_of_edges())

        hl = HubLabeling(G, max_hubs=10)

```
